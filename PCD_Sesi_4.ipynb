```python
import numpy as np
import imageio
from scipy.ndimage import gaussian_filter
import matplotlib.pyplot as plt

# Function to process the image
def process_image(input_path, output_path, sigma):
    # Read the image
    image = imageio.imread(input_path)

    # Convert image to grayscale if it's colored
    if image.ndim == 3:
        image = np.mean(image, axis=2).astype(np.uint8)

    # Apply Gaussian filter
    smoothed_image = gaussian_filter(image, sigma=sigma)

    # Save the processed image
    imageio.imwrite(output_path, smoothed_image)

    # Display original and processed images
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(image, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Image after Gaussian Filter")
    plt.imshow(smoothed_image, cmap='gray')
    plt.axis('off')

    plt.show()

# Paths for input and output images
input_image_path = '/content/Gunung.jpg'# Change to your input image path
output_image_path = '/content/Pantai.jpg'# Change to your output image path

# Gaussian filter sigma value
sigma_value = 2

# Run the function
process_image(input_image_path, output_image_path, sigma_value)
```
